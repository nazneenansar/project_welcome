```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LSTM
from keras.utils import np_utils

archives = []
tweets = []
```

```python
import csv
file = open("csv_file.csv")
csvreader = csv.reader(file)

for line in csvreader:
    print(line)
    words = [x for x in line[0].replace("\n", " ").replace("&amp;", " and ").split(" ") if len(x) > 0]
    words = [x for x in words if x.isalnum() or (x.find('#') < 0 and 
                       x.find('@') < 0 and x.find('/') < 0) and x.isascii()]
    if words not in archives:
      archives.append(words)
      tweets.append(' '.join(words))
print(tweets)      
file.close()
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
happy_tweets = []

for tweet in tweets:
  if sid.polarity_scores(tweet).get('pos') > sid.polarity_scores(tweet).get('neg'):
    happy_tweets.append(tweet)
print(happy_tweets)
print(len(happy_tweets))
print(len(tweets))
```

```python
with open('final-tweets.txt', 'w') as f:
    for tweet in happy_tweets:
        f.write(tweet)
        f.write('\n')
```
